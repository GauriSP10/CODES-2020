```python
#Importing necessary module.
from PIL import Image

#Reading the Image.
img = Image.open('Simpsons.jpg')

#Displaying the Image.
img.show()
```

```python
#Importing necessary modules.
from PIL import Image
import matplotlib.pyplot as plt

#Reading the Image.
img = plt.imread('Simpsons.jpg')

#Displaying the Image as a plot.
plt.imshow(img)
```

```python
#Importing necessary modules.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#Reading the Image.
img = mpimg.imread('Simpsons.jpg')

#Displaying the Image as a plot.
plt.imshow(img)
```

```python
#Importing necessary modules.
import imageio
import matplotlib.pyplot as plt

#Reading the Image.
img = imageio.imread('Simpsons.jpg')

#Displaying the Image as a plot.
plt.imshow(img)
```

```python
#Importing necessary module.
import cv2 as cv2

#Reading the Image.
img = cv2.imread('Simpsons.jpg',1)

#Displaying the Image as a plot.
cv2.imshow('Title', img)

#Displaying the Image until a key is pressed.
cv2.waitKey(0)
```

```python
#Importing necessary modules.
import cv2 as cv2
import matplotlib.pyplot as plt

#Reading the Image in grayscale.
img = cv2.imread('Simpsons.jpg',-1)

#Converting GBR colour mode to RGB colour mode.
RGBimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Displaying the Image as a plot.
plt.imshow(RGBimg)
```
