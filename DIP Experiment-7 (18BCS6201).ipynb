<b> Image blurring and filtering </b>

<b> Blurring image </b>

```python
#Importing necessary modules.
import cv2

#Reading the image.
img = cv2.imread('c.png')

#Kernel size can be changed as wanted.
blurImg = cv2.blur(img,(10,10))

#Returning the image.
cv2.imshow('Blurred Image',blurImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
#There is an image blurring technique called 'Averaging'.
```

<b> Gaussian Blurring, Median Blurring and Bilateral Filtering </b>

```python
#Importing necessary modules.
import cv2

#Reading the image.
img = cv2.imread('c.png')

#Kernel size can be changed as wanted.
avging = cv2.blur(img,(10,10))

#Returning the image.
cv2.imshow('Averaging',avging)
cv2.waitKey(0)

#Gaussian blurring.
#Kernel size can be changed as wanted.
gausBlur = cv2.GaussianBlur(img, (5,5),0)
#Returning the image.
cv2.imshow('Gaussian Blur', gausBlur)
cv2.waitKey(0)

#Median blurring.
medBlur = cv2.medianBlur(img,5)
#Returning the image.
cv2.imshow('Median Blur', medBlur)
cv2.waitKey(0)

#Bilateral filtering.
bilFilter = cv2.bilateralFilter(img,9,75,75)
#Returning the image.
cv2.imshow('Bilateral Filtering', bilFilter)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<b> Motion Blur </b>

```python
#Importing necessary modules.
import cv2
import numpy as np

#Reading the image.
img = cv2.imread('c.png')

#Specifying the kernel size.
#The greater the kernel size, the more the motion is.
kernel_size = 30

#Creating the vertical kernel.
kernel_v = np.zeros((kernel_size, kernel_size))

#Creating a copy of the same for creating the horizontal kernel.
kernel_h = np.copy(kernel_v)

#Filling the middle row with ones.
kernel_v[:, int((kernel_size - 1)/2)] = np.ones(kernel_size)
kernel_h[int((kernel_size - 1)/2), :] = np.ones(kernel_size)

# Normalizing.
kernel_v /= kernel_size
kernel_h /= kernel_size

#Applying the vertical kernel.
vertical_mb = cv2.filter2D(img, -1, kernel_v)

# Applying the horizontal kernel.
horizonal_mb = cv2.filter2D(img, -1, kernel_h)

#Saving the returned images.
cv2.imwrite('a_vertical.png', vertical_mb)
cv2.imwrite('a_horizontal.png', horizonal_mb)

#Returning the images.
cv2.imshow('a_vertical', vertical_mb)
cv2.imshow('a_horizontal', horizonal_mb)
```
