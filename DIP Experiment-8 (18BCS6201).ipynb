<b> Image Histogram, Image Pyramid and Image Translation </b>

<b> Image Histogram </b>

```python
#Importing necessary modules.
import cv2
from matplotlib import pyplot as plt

#Reading the image.
img = cv2.imread('PrisonBreak.jpg',0)

#Finding frequency of the pixels in the range '0-255'.
histr = cv2.calcHist([img],[0],None,[256],[0,255])

#Returning the plot of the image.
plt.plot(histr)
plt.show()
```

```python
#Importing necessary modules.
import cv2
from matplotlib import pyplot as plt

#Reading the image.
img = cv2.imread('PrisonBreak.jpg',2)

#Returning the plot of the image,alternate way.
plt.hist(img.ravel(),256,[0,255])
plt.show()
```

<b> Image Pyramid </b>

```python
#Importing necessary modules.
import cv2
import matplotlib.pyplot as plt

#Reading the image.
img = cv2.imread("PrisonBreak.jpg")
layer = img.copy()

for i in range(4):
    plt.subplot(2, 2, i + 1)
    
#Implementing the function 'pyrDown()'.
layer = cv2.pyrDown(layer)

#Returning the plot of the image.
plt.imshow(layer)
cv2.imshow("str(i)", layer)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<b> Image Translation </b>

```python
#Importing necessary modules.
import cv2
import numpy as np

#Reading the image.
image = cv2.imread('PrisonBreak.jpg')

#Storing the height and width of the image.
height, width = image.shape[:2]
quarter_height, quarter_width = height / 4, width / 4
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

#Using 'warpAffine()' to transform the image the image using the matrix:'T'.
img_translation = cv2.warpAffine(image, T, (width, height))

#Returning the image.
cv2.imshow("Original Image", image)
cv2.imshow('Translation', img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```
