```python
#Importing necessary modules.
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv2

#Reading the Image in grayscale.
img = cv2.imread('Dark.jpg', cv2.IMREAD_GRAYSCALE)

#Displaying the Image.
plt.imshow(img, cmap='gray')
plt.show()
```

```python
#Converting GBR colour mode to RGB colour mode.
RGBimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Plotting on the Image.
plt.plot([100,600] , [200,600] , 'm' , linewidth=6)

#Displaying the Image.
plt.imshow(RGBimg)

#Saving the Image.
cv2.imwrite('Darker.png' , img)
```

```python
#Importing necessary modules.
import numpy as np
import cv2 as cv2

#Capturing the Video.
cap = cv2.VideoCapture(r'C:\Users\Helenasec\Dark.mp4')
 
while(True):
    
    #Putting infinite loop,ret_repeat
    ret, frame = cap.read()
    
    #Displaying the Video.
    cv2.imshow('Window',frame)
    
    #Breaking later by a break statement.
    if cv2.waitKey(1) & 0xFF == ord('1'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
#Importing necessary modules.
import numpy as np
import cv2 as cv2

#Capturing the Video.
cap = cv2.VideoCapture('Dark.mp4')
 
while(True):
    
    #Putting infinite loop,ret_repeat
    ret, frame = cap.read()
    
    #Converting to grayscale.
    grayimg = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #Displaying the Video.
    cv2.imshow('Frame Window',frame)
    cv2.imshow('Gray Window', grayimg)
    
    #Breaking later by a break statement.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
