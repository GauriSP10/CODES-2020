```python
#Importing necessary modules.
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

#Reading the image.
img = imageio.imread('a.jpg')

#Plotting the image.
plt.figure(figsize = (5,5))
plt.imshow(img)
print('Type of the image : ' , type(img))
print('Shape of the image : {}'.format(img.shape))

#The shape of the ndarray shows that it is a three-layered matrix.
print('Image Hight {}'.format(img.shape[0]))
print('Image Width {}'.format(img.shape[1]))
print('Dimension of Image {}'.format(img.ndim))

#The 'ndim()' function return the number of dimensions of an array.
print('Image size {}'.format(img.size))
print('Maximum RGB value in this image {}'.format(img.max()))
print('Minimum RGB value in this image {}'.format(img.min()))

#Locating a specific pixel at Row : 120 ; Column : 50
#Returning each channel's value.
print('Value of only R channel {}'.format(img[ 120, 50, 0]))
print('Value of only G channel {}'.format(img[ 120, 50, 1]))
print('Value of only B channel {}'.format(img[ 120, 50, 2]))
```

<b> Displaying each channel </b>

```python
#Importing necessary modules.
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

#Reading the image.
img = imageio.imread('a.jpg')

#Returning 'R' specific channel.
plt.title('Channel R')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))
plt.imshow(img[ : , : , 0])
plt.show()
```

```python
#Returning 'G' specific channel.
plt.title('Channel G')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))
plt.imshow(img[ : , : , 1])
plt.show()
```

```python
#Returning 'B' specific channel.
plt.title('Channel B')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))
plt.imshow(img[ : , : , 2])
plt.show()
```

<b> Changing RGB values </b>

```python
#Importing necessary modules.
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

#Reading the image.
img = imageio.imread('a.jpg')

#Maximising the intensity of the pixel's of 'R' channel.
img[50:100 , : , 0] = 255
plt.figure( figsize = (9,6))
plt.imshow(img)
plt.show()
```

```python
#Maximising the intensity of the pixel's of 'G' channel.
img[150:200 , : , 1] = 255
plt.figure( figsize = (10,6))
plt.imshow(img)
plt.show()
```

```python
#Maximising the intensity of the pixel's of 'B' channel.
img[250:300 , : , 2] = 255
plt.figure( figsize = (9,6))
plt.imshow(img)
plt.show()
```

```python
#Setting value as 255 for all channels to turn those pixels to white colour.
img[ 50:250 , 300:400 , [0,1,2] ] = 255
plt.figure( figsize = (5,5))
plt.imshow(img)
plt.show()
```

<b> Splitting Layers </b>

```python
#Importing necessary modules.
import numpy as np
pic = imageio.imread('a.jpg')
fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5))
for c, ax in zip(range(3), ax):
#Creating zero matrix.
    split_img = np.zeros(pic.shape, dtype="uint8")
#'dtype' by default is 'numpy.float64'.
    split_img[ :, :, c] = pic[ :, :, c]
    
#Returning each channel.
    ax.imshow(split_img)
```

<b> Threshold using OpenCv </b>

```python
#Importing necessary modules.
import cv2
import numpy as np

#Reading the image.
img = cv2.imread('a.jpg')

#Thresholding.
retval, threshold = cv2.threshold(img, 12, 255, cv2.THRESH_BINARY)
cv2.imshow('Original',img)
cv2.imshow('Thresholded',threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<b> Thresholding an image after converting it to grayscale image </b>

```python
#Importing necessary modules.
import cv2
import numpy as np

#Converting the image to grayscale.
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 10, 255, cv2.THRESH_BINARY)
cv2.imshow('Original',img)
cv2.imshow('Thresholded',threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
