<b> Performing operations on the dataframe </b>

```python
#Defining function 'missingfromframe'.
def missingfromframe(DataF):
        # Returning number of missing values.
        Value = DataF.isnull().sum()
        
        # Returning percentage of missing values.
        Percentage = 100 * DataF.isnull().sum() / len(Data)
        
        # Returning a fresh table with results.
        Table = pd.concat([Value, Percentage], axis=1)
        
        # Renaming the columns.
        TableRenamed = Table.rename(columns = {0 : 'No. of  Values Missing ', 1 : '% of the Total Values'})
        
        # Sorting the table by percentage of missing values in descending order.
        TableRenamed = TableRenamed[TableRenamed.iloc[:,1] != 0].sort_values('% of the Total Values', ascending=False).round(1)
        
        # Return the dataframe with missing information
        return TableRenamed
```

```python
#Passing dataframe 'Data' into function 'missingfromframe' to obtain the following columns.
missingfromframe(Data)
```

```python
#Returning the above output to a dataframe named 'missingfromframe_df'
missingfromframe_df = missingfromframe(Data)

#Altering/Removing columns having missing data greater than 10%
missingfromframe_columns = list(missingfromframe_df[missingfromframe_df['% of the Total Values'] > 10].index)

#Dropping columns.
DataFnew = Data.drop(columns = list(missingfromframe_columns))
```

```python
#Returning the freshly created dataframe 'DataFnew' by filling in the missing values with the mean of the corresponding column.
DataFnew['PctEmployed16_Over']=DataFnew['PctEmployed16_Over'].fillna(DataFnew['PctEmployed16_Over'].mean())
DataFnew
```

<b> Building Model </b>

```python
#Importing libraries and modules.
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Splitting 'DataFnew' into 2 subsets for training and testing using train_test_split function.
train_a , test_a = train_test_split(DataFnew , train_size = 0.7 ,test_size = 0.3 , random_state = 123)
```

```python
#Returning columns for 'train_a'
train_a.columns
```

```python
#Creating dataframe 'a' by dropping attributes from 'DataFnew'
a = DataFnew.drop(['TARGET_deathRate' , 'PctMarriedHouseholds' , 'popEst2015' , 'BirthRate' , 'PctBachDeg25_Over' , 
                   'PctPublicCoverage' ,'PctBachDeg18_24', 'PctPrivateCoverage' , 'MedianAgeMale' ,  'medIncome' ,
                   'PctUnemployed16_Over' , 'avgDeathsPerYear' ,  'AvgHouseholdSize','PctWhite'   ] , axis =1)


#Creating dataframe 'b' with only 'TARGET_deathRate' attribute.
b = DataFnew['TARGET_deathRate']
```

```python
#Assigning 'Lr'.
Lr = LinearRegression()

#Taking training data as arguments using .fit(,) function.
Lr.fit(a,b)

Rfe = RFE(Lr,10)
Rfe =Rfe.fit(a,b)
```

```python
#Returning list 'ls' and applying 'zip()' function to it to return pairs of values.
ls=list(zip(a.columns , Rfe.support_ ,Rfe.ranking_))
ls
```

```python
#Returning columns.
cols = a.columns[Rfe.support_]
cols
```

```python
a.columns[~Rfe.support_]
```

```python
a_Rfe = a[cols]
```

```python
#Importing necessary library.
import statsmodels.api as sm

#Performing Regression and returning summary.
#a_Rfe = sm.add_constant(a_Rfe)
Lr = sm.OLS(b,a_Rfe).fit()
print(Lr.summary())
```

```python
#Importing necessary library.
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Variance Inflation Factor
Vif = pd.DataFrame()
c = a_Rfe
Vif['features'] = c.columns
Vif['Vif'] = [variance_inflation_factor(c.values , i) for i in range(c.shape[1])]
Vif['Vif'] = round(Vif['Vif'] , 2)

#Sorting dataframe 'Vif'
Vif = Vif.sort_values(by ='Vif' , ascending = False)
Vif
```

```python
Priceb = Lr.predict(a_Rfe)

#Plotting graph.
figure = plt.figure()
sns.distplot((b - Priceb ),bins = 20)
figure.suptitle('Terms of Error' , fontsize=20)
plt.xlabel('Errors!' , fontsize = 18)
```
